```python
import multiprocessing

def sum_nums(low,high):
    result = 0
    for i in range(low,high+1):
        result += i
    return result

if __name__ == "__main__":
    n = 1000 
    procs = 2

    sizeSegment = n/procs

    jobs = []
    for i in range(0, procs):
        process = multiprocessing.Process(target=sum_nums, args=(i*sizeSegment+1, (i+1)*sizeSegment))
        jobs.append(process)

    for j in jobs:
        j.start()
    for j in jobs:
        j.join()
```

Crear una lista de n elementos $[x_1,x_2,\ldots x_{10}]$, y elevarlos al cuadrado $[x_1^2, x_2^2, \ldots, x_n^2]$

```python
nn = range(10)
nn = list(nn)
print(nn)
```

```python
# SOLUCIÃN 1
for i in nn: #recorrido de la estructura
  # elento i de la lista tiene tal orden
  nn[i] = nn[i]**2
print(nn)
```

```python
# SOLUCIÃN 2
e = []
for num in nn:
  e.append(num**2)
```

```python
print(e)
```

```python
e.pop(3)
print(e)
```
